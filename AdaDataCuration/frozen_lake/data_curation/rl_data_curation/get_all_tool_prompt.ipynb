```python
input_sft_data = "/mnt/petrelfs/songmingyang/code/reasoning/tool_util_code/frozen_lake/data_curation/frozen_lake_metadata_v2/merged_sft_file/vsp2tasks_v2_navigationa.json"
input_sft_data = load_json_file(input_sft_data)
len(input_sft_data)
```

```python
input_sft_data[1]
```

```python
# First let's see the structure of the conversations in one entry
example_conversations = input_sft_data[0]['conversations']
print(f"Original first conversation structure: {example_conversations[0]}")

# Now let's update all entries in the input_sft_data list
for item in input_sft_data:
    if 'conversations' in item and len(item['conversations']) > 0:
        # Assuming the first element in conversations is the system prompt
        if item['conversations'][0].get('from') == 'system':
            # Update the system prompt with our new prompt
            item['conversations'][0]['value'] = prompt

# Verify the change with the first item
print(f"Updated first conversation: {input_sft_data[0]['conversations'][0]}")

# Count how many items were updated
updated_count = sum(1 for item in input_sft_data 
                   if 'conversations' in item 
                   and len(item['conversations']) > 0 
                   and item['conversations'][0].get('from') == 'system')

print(f"Updated {updated_count} items out of {len(input_sft_data)}")
```

```python
def read_image_to_bytes(image_path):
    """读取图片文件并转换为bytes格式"""
    try:
        with open(image_path, 'rb') as f:
            return f.read()
    except Exception as e:
        print(f"读取图片失败: {image_path}, 错误: {e}")
        return None
```

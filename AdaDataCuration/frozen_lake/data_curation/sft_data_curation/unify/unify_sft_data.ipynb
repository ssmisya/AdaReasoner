```python
from tool_server.utils.utils import *
import os
import random

random.seed(42)
```

```python
jigsaw_sft_data = "/mnt/petrelfs/share_data/songmingyang/data/vl_reasoning/tool_dataset/jigsaw/jigsaw_metadata_v1/splits/sft/gemini_enhanced/sft_data/jigsaw_sft_v1.json"

vsp_sft_data = "/mnt/petrelfs/songmingyang/code/reasoning/tool_util_code/frozen_lake/data_curation/frozen_lake_metadata_v2/merged_sft_file/vsp2tasks_v2_navigationa.json"

web_sft_data = "/mnt/petrelfs/share_data/sunhaoyu/datasets/web/webdata_ocr_sharegpt.json"

jigsaw_sft_data = "/mnt/petrelfs/songmingyang/code/reasoning/tool_util_code/frozen_lake/data_curation/frozen_lake_metadata_v2/merged_sft_file/randomized/jigsaw_sft_v1_randomized.json"

vsp_sft_data = "/mnt/petrelfs/songmingyang/code/reasoning/tool_util_code/frozen_lake/data_curation/frozen_lake_metadata_v2/merged_sft_file/randomized/vsp2tasks_v2_navigationa_new_desc.json"

web_sft_data = "/mnt/petrelfs/songmingyang/code/reasoning/tool_util_code/frozen_lake/data_curation/frozen_lake_metadata_v2/merged_sft_file/randomized/web_v1_new_desc.json"

jigsaw_sft_data = load_json_file(jigsaw_sft_data)
vsp_sft_data = load_json_file(vsp_sft_data)
web_sft_data = load_json_file(web_sft_data)

print(f"Length of jigsaw_sft_data: {len(jigsaw_sft_data)}, vsp_sft_data: {len(vsp_sft_data)}, web_sft_data: {len(web_sft_data)}")
```

```python
new_web_sft_data = []
for item in web_sft_data:
    images = item["images"]
    non_image_flag = False
    for image_path in images:
        if not os.path.exists(image_path):
            non_image_flag = True
            break
   
        
    conversation_str = str(item["conversations"])
    convesation_length = len(conversation_str)
    if convesation_length > 35536:
        non_image_flag = True
        
    if not non_image_flag:
        new_web_sft_data.append(item)

print(f"Length of filtered web_sft_data: {len(new_web_sft_data)}")
web_sft_data = new_web_sft_data
```

```python
common_volumn = min(len(jigsaw_sft_data), len(vsp_sft_data), len(web_sft_data))
merged_sft_data = []
random.shuffle(jigsaw_sft_data)
random.shuffle(vsp_sft_data)
random.shuffle(web_sft_data)
for i in range(common_volumn):
    merged_sft_data.append(jigsaw_sft_data[i])
    merged_sft_data.append(vsp_sft_data[i])
    merged_sft_data.append(web_sft_data[i])

random.shuffle(merged_sft_data)

output_path = "/mnt/petrelfs/songmingyang/songmingyang/data/vl_reasoning/tool_dataset/unified_tool/sft_data/merged_sft_jigsaw_vsp_web_v1_randomized.json"
```

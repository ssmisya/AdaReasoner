```python
from tool_server.utils.utils import *
from tool_server.utils.prompts import tool_planning_model_prompt_one_tool_call
from tqdm import tqdm
import json, os
refocus_data1="/mnt/petrelfs/songmingyang/code/reasoning/data_construction/refocus/data/sft_data/refocus_chartqa_v_bar_wbb_reformatted.json"
refocus_data2="/mnt/petrelfs/songmingyang/code/reasoning/data_construction/refocus/data/sft_data/refocus_chartqa_v_bar_wbb_selfbar_candidates.json"
pixel_reasoner_data1="/mnt/petrelfs/songmingyang/code/reasoning/data_construction/PixelReasoner/data/sft_data/pixelreasonersft_groundingcrop.json"
pixel_reasoner_data2="/mnt/petrelfs/songmingyang/code/reasoning/data_construction/PixelReasoner/data/sft_data/pixelreasonersft_crop.json"
data_list = [refocus_data1, refocus_data2, pixel_reasoner_data1, pixel_reasoner_data2]

output_path = "/mnt/petrelfs/songmingyang/code/reasoning/data_construction/combine/data/toolsft_v1.json"
```

```python
final_list = []
for data_path in data_list:
    current_data = load_json_file(data_path)
    for item in current_data:
        images = item["images"]
        for image_path in images:
            assert os.path.exists(image_path), f"Image path {image_path} does not exist."
        conversations = item["conversations"]
        
        image_token_num = 0
        for conv in conversations:
            assert conv["from"] in ["human", "gpt", "system"]
            conv_value = conv["value"]
            count = conv_value.count("<image>")
            image_token_num += count
            
        assert len(images) == image_token_num, f"Image token count mismatch in {data_path} for item {item['qid']}"
        final_list.append(item)
        
```

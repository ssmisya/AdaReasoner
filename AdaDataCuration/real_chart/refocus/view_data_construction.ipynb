```python
import os
from mr_eval.utils.utils import *
from tqdm import tqdm

# refocus_data_path = "/mnt/petrelfs/songmingyang/songmingyang/data/vl_reasoning/ReFocus_data/train_chartQA_h_bar_wbb/gpt-4o-2024-08-06-mix_orig_build_train_nogold"
refocus_data_path = "/mnt/petrelfs/songmingyang/songmingyang/data/vl_reasoning/ReFocus_data/train_chartQA_v_bar_wbb/gpt-4o-2024-08-06-mix_orig_build_train_nogold"
# refocus_data_path = "/mnt/petrelfs/songmingyang/songmingyang/data/vl_reasoning/ReFocus_data/train_chartQA_h_bar_wbb-nogold_errors/gpt-4o-2024-08-06-mix_orig_build_train"
# refocus_data_path = "/mnt/petrelfs/songmingyang/songmingyang/data/vl_reasoning/ReFocus_data/train_chartQA_v_bar_wbb-nogold_errors/gpt-4o-2024-08-06-mix_orig_build_train"

```

```python
refocus_instances = os.listdir(refocus_data_path)
total_num = len(refocus_instances)
action_dict = {}
```

```python
for i, instance in enumerate(tqdm(refocus_instances)):
    current_instance_path = os.path.join(refocus_data_path, instance)
    current_answer_json = os.path.join(current_instance_path, "answer.json")
    answer_file = load_json_file(current_answer_json)
    answer_str = answer_file["answer"]
    assert isinstance(answer_str, str), f"Answer for {instance} is not a string"
    if "ACTION 0:" not in answer_str:
        action_str = "FORMAT ERROR"
    else:
        action_str = answer_str.split("ACTION 0:")[1].split("ANSWER:")[0].strip()
    if action_str not in action_dict:
        action_dict[action_str] = 0
    action_dict[action_str] += 1
```

```python
action_dict_list = sorted(action_dict.items(), key=lambda x: x[1], reverse=True)
action_dict_list
```

```python
tool_call_items = [item for item in action_dict_list if "FORMAT ERROR" not in item[0] and "No action needed." not in item[0]]
```

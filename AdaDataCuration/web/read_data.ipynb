```python
import json

def read_first_n_lines_jsonl(file_path, n):
    """
    读取并解析一个 JSONL 文件的前 n 行。

    参数:
    file_path (str): JSONL 文件的路径。
    n (int): 需要读取的行数。

    返回:
    list: 一个包含已解析的JSON对象（字典）的列表。
    """
    data = []
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                # 如果已经读取了 n 行，则停止循环
                if i >= n:
                    break
                # 解析当前行并添加到列表中
                data.append(json.loads(line.strip()))
    except FileNotFoundError:
        print(f"错误：文件 '{file_path}' 未找到。")
    except json.JSONDecodeError as e:
        # 如果某一行不是有效的JSON，会捕获错误
        print(f"错误：解析第 {i+1} 行时发生JSON解码错误: {e}")
    except Exception as e:
        print(f"发生未知错误: {e}")
        
    return data




# 2. 设置文件路径和要读取的行数
file_path = 'AdaReasoner/AdaDataCurationweb/test_results/guichat/72b_responses.jsonl'
num_lines_to_read = 10

# 3. 调用函数读取文件
first_lines = read_first_n_lines_jsonl(file_path, num_lines_to_read)

# 4. 打印结果
if first_lines:
    print(f"成功读取文件 '{file_path}' 的前 {len(first_lines)} 行内容：")
    for item in first_lines:
        print(item)
```

```python
import json

def filter_jsonl_by_score(input_file_path, output_file_path, score_to_keep=0):
    """
    读取一个 JSONL 文件，筛选出 'llm_score' 为特定值的数据，
    并将其保存到新的 JSONL 文件中。

    参数:
    input_file_path (str): 输入的 JSONL 文件路径。
    output_file_path (str): 输出的 JSONL 文件路径。
    score_to_keep (int): 需要保留的数据的 llm_score 值，默认为 0。
    """
    lines_read = 0
    lines_written = 0
    
    try:
        # 使用 'with' 语句可以安全地打开和关闭文件
        # 同时打开输入和输出文件
        with open(input_file_path, 'r', encoding='utf-8') as infile, \
             open(output_file_path, 'w', encoding='utf-8') as outfile:
            
            # 逐行读取输入文件
            for line in infile:
                lines_read += 1
                try:
                    # 将单行JSON字符串解析为Python字典
                    data = json.loads(line)
                    
                    # 检查 'llm_score' 字段是否存在且值是否为 0
                    # 使用 .get() 方法比直接用 [] 更安全，如果键不存在不会报错
                    if data.get('llm_score') == score_to_keep:
                        # 将符合条件的字典转换回JSON字符串并写入新文件
                        # ensure_ascii=False 保证中文字符正常写入
                        outfile.write(json.dumps(data, ensure_ascii=False) + '\n')
                        lines_written += 1
                        
                except json.JSONDecodeError:
                    print(f"警告：第 {lines_read} 行不是有效的JSON，已跳过。")
                except Exception as e:
                    print(f"处理第 {lines_read} 行时发生错误: {e}")

        print("\n处理完成！")
        print(f"总共读取了 {lines_read} 行数据。")
        print(f"成功将 {lines_written} 行 'llm_score' 为 {score_to_keep} 的数据保存到 '{output_file_path}'。")

    except FileNotFoundError:
        print(f"错误：输入文件 '{input_file_path}' 未找到。")
    except Exception as e:
        print(f"发生未知错误: {e}")



def filter_jsonl_by_score(input_file_path, output_file_path, score_to_keep=0):
    """
    读取一个 JSONL 文件，筛选出 'llm_score' 为特定值的数据，
    并将其保存到新的 JSONL 文件中。

    参数:
    input_file_path (str): 输入的 JSONL 文件路径。
    output_file_path (str): 输出的 JSONL 文件路径。
    score_to_keep (int): 需要保留的数据的 llm_score 值，默认为 0。
    """
    lines_read = 0
    lines_written = 0
    
    try:
        # 使用 'with' 语句可以安全地打开和关闭文件
        # 同时打开输入和输出文件
        with open(input_file_path, 'r', encoding='utf-8') as infile, \
             open(output_file_path, 'w', encoding='utf-8') as outfile:
            
            # 逐行读取输入文件
            for line in infile:
                lines_read += 1
                try:
                    # 将单行JSON字符串解析为Python字典
                    data = json.loads(line)
                    
                    # 检查 'llm_score' 字段是否存在且值是否为 0
                    # 使用 .get() 方法比直接用 [] 更安全，如果键不存在不会报错
                    if data.get('llm_score') == score_to_keep:
                        # 将符合条件的字典转换回JSON字符串并写入新文件
                        # ensure_ascii=False 保证中文字符正常写入
                        outfile.write(json.dumps(data, ensure_ascii=False) + '\n')
                        lines_written += 1
                        
                except json.JSONDecodeError:
                    print(f"警告：第 {lines_read} 行不是有效的JSON，已跳过。")
                except Exception as e:
                    print(f"处理第 {lines_read} 行时发生错误: {e}")

        print("\n处理完成！")
        print(f"总共读取了 {lines_read} 行数据。")
        print(f"成功将 {lines_written} 行 'llm_score' 为 {score_to_keep} 的数据保存到 '{output_file_path}'。")

    except FileNotFoundError:
        print(f"错误：输入文件 '{input_file_path}' 未找到。")
    except Exception as e:
        print(f"发生未知错误: {e}")

input_filename = 'AdaReasoner/AdaDataCurationweb/test_results/7b_full_results_llm.jsonl'
output_filename = 'AdaReasoner/AdaDataCurationweb/test_results/7b_full_results_llm_wrong.jsonl'

# 2. 调用函数执行筛选和保存操作
filter_jsonl_by_score(input_filename, output_filename)
```
